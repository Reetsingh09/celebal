```python
# Load the CSV file into a DataFrame
df = pd.read_csv('sales_data.csv')

print("✅ Dataset loaded successfully!")
print(f"   → {df.shape[0]} rows  ×  {df.shape[1]} columns")
```

```python
# Count NaN per column
missing = df.isnull().sum()
print("Missing values per column:")
print(missing)
print(f"\nTotal missing cells: {missing.sum()}")
```

```python
df.isnull()
```

```python
# Modern Pandas (CoW-safe) — avoid chained assignment
df['customer_name'] = df['customer_name'].fillna('Unknown Customer')
df['price']         = df['price'].fillna(df['price'].median())
df['quantity']      = df['quantity'].fillna(1)
print("✅ Missing values handled!")
print(df.isnull().sum())
```

# New Section

```python
electronics_df = df[df['category'] == 'Electronics']
print(f"Electronics orders: {len(electronics_df)}")
electronics_df.head()
```

```python
bulk_orders = df[df['quantity'] > 2]
print(f"Bulk orders (qty > 2): {len(bulk_orders)}")
bulk_orders[['order_id', 'customer_name', 'product', 'quantity']]
```

```python
key_cols = df[['order_id', 'customer_name', 'product', 'price', 'quantity']]
key_cols.head()
```

```python
north_electronics = df[(df['category'] == 'Electronics') & (df['region'] == 'North')]
print(f"Electronics in North: {len(north_electronics)}")
north_electronics[['order_id', 'customer_name', 'product', 'price', 'region']]
```

```python
before = len(df)
print(f"Rows BEFORE removing duplicates: {before}")
# Show duplicate rows
dupes = df[df.duplicated()]
print(f"\nDuplicate rows found: {len(dupes)}")
dupes
```

```python
df = df.drop_duplicates()
after = len(df)
print(f"Rows AFTER removing duplicates : {after}")
print(f"Duplicates removed             : {before - after}")
```

```python
df['total_amount'] = df['price'] * df['quantity']
print("✅ Column 'total_amount' created!")
df[['order_id', 'product', 'price', 'quantity', 'total_amount']].head(8)
```

```python
summary = df.groupby('category')['total_amount'].agg(
    total_revenue='sum',
    avg_order_value='mean',
    num_orders='count'
).round(2)
print(summary)
```

```python
region_summary = df.groupby('region')['total_amount'].sum().sort_values(ascending=False)
print(region_summary)
```

```python
output_path = 'cleaned_sales_data_reet.csv'
df.to_csv(output_path, index=False)
print(f"✅ Cleaned dataset saved → '{output_path}'")
print(f"   Rows    : {df.shape[0]}")
print(f"   Columns : {df.shape[1]}")
print(f"   Columns : {list(df.columns)}")
```
